## Boxplot

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="white", context="talk")
```

```python
rs = np.random.RandomState(8)
#Number of random state are 8
```

```python
x1 = np.array(list("ABCDEFGHIJ"))
```

```python
y1 =  np.arange(1,11)
```

```python
sns.barplot(x=x1,y=y1 , palette='rocket').axhline(0, color="k", clip_on=False);
```

```python
y2 = y1-5.5
sns.barplot(x=x1,y=y2 , palette='vlag').axhline(0, color="k", clip_on=False);
# axhline(0, color="k", clip_on=False)  make a line on 0  ,  if you chose 10 then line on number 10 
```

```python
y3 = rs.choice(y1, len(y1), replace=False)
# random choice 
sns.barplot(x=x1, y=y3, palette="deep")
sns.despine(bottom=True)  # get rid of line  
```

**Question**
: What is the relationship between 'comb' and 'VClass'?. The cars in this dataset are categorized into one of five different vehicle classes based on size. Starting from the smallest, they are: {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, and Large Cars}.

```python
df = pd.read_csv('fuel_econ.csv')
```

```python
sedan_classess = ['Minicompact Cars','Subcompact Cars','Compact Cars','Midsize Cars','Large Cars']
v_class = pd.api.types.CategoricalDtype(ordered=True,categories=sedan_classess) 
#Type for categorical data with the categories and orderedness
df['VClass'] = df['VClass'].astype(v_class);
```

```python
base_color = sns.color_palette()[0]
sns.barplot(data=df,x='VClass',y='displ',color=base_color)
plt.xticks(rotation=15); # or vertical 
```

## Advance: Boxenplot

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
diamonds.info()
```

```python
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.boxenplot(data=diamonds,x="clarity",y="carat",
              scale="linear",order=clarity_ranking,color="b")
# diamond shape in the table is the ideal data of each clarities
```

```python
sns.boxplot(data=diamonds,x="clarity",y="carat",order=clarity_ranking
           ,color="b")
```

**From boxplot it's hart to figure out the distribution of all data
