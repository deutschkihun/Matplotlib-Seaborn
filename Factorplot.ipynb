```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('train.csv')
# titanic dataset
```

```python
df.head()
```

```python
sns.factorplot('Pclass','Survived',hue='Sex',data=df)
plt.title('Pclass & Survivied relationship');
# Pclass = Ticket class
```

```python
sns.factorplot('Pclass','Survived',hue='Sex',col='Embarked',data=df);
# Pclass & Survivied & Embarkation relationship
# Port of Embarkation / C = Cherbourg, Q = Queenstown, S = Southampton
```

```python
sns.factorplot('Embarked','Survived',hue='Sex',data=df)
plt.title('Embarkation & Survivied relationship');
# Port of Embarkation / C = Cherbourg, Q = Queenstown, S = Southampton
```

```python
sns.factorplot('SibSp','Survived',hue='Sex',data=df)
plt.title('SipSp & Survivied relationship');
# SipSp: of siblings / spouses aboard the Titanic
```

```python
sns.factorplot('Parch','Survived',hue='Sex',data=df)
plt.title('SipSp & Survivied relationship');
# of parents / children aboard the Titanic
```

```python
#use name feature
df['original'] = 0 
for i in df:
    df['original']=df.Name.str.extract('([A-Za-z]+)\.')
    # extract all alpabet combination with 'period' -> tells us men or women
```

```python
df['original'].value_counts()
```

```python
#Replace salutation to make consistent 
df['original'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],
['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],inplace=True)
```
